```python
#Data Loading

import pandas as pd

df = pd.read_csv("cs-training.csv")
df.head()
```

```python
#Data Exploration

df.info()
df.describe()
df.isnull().sum()
```

```python
# First Data Cleaning

# Eliminate Index Columns
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# Nuls
df.isnull().sum()
```

```python
# Handling Missing Values

# MonthlyIncome → median imputation (robust to outliers)
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)

# NumberOfDependents → mode imputation (most common value)
df['NumberOfDependents'].fillna(df['NumberOfDependents'].mode()[0], inplace=True)

# Verify missing values
df.isnull().sum()
```

```python
# Outlier Detection

# Summary statistics to identify extreme values
df.describe()
```

## Outlier Treatment

Extreme values were identified and removed based on business logic and statistical thresholds to ensure realistic data.

```python
# Store original dataset shape before outlier treatment

original_shape = df.shape
original_shape
```

```python
# Outlier Treatment

# Remove unrealistic ages
df = df[df['age'] > 18]

# Limit extreme DebtRatio values
df = df[df['DebtRatio'] < 5]

# Limit credit utilization to realistic range (<= 1)
df = df[df['RevolvingUtilizationOfUnsecuredLines'] <= 1]

# Remove extreme income outliers (top 1%)
df = df[df['MonthlyIncome'] < df['MonthlyIncome'].quantile(0.99)]
```

```python
# Final Data Validation

cleaned_shape = df.shape

print("Before cleaning:", original_shape)
print("After cleaning:", cleaned_shape)

df.describe()
```

## Data Reduction

After applying outlier filtering, the dataset was reduced from 150,000 to 116,182 observations.

Approximately 22% of the data was removed due to unrealistic or extreme values. This step ensures higher data quality and more reliable analysis by focusing on realistic financial behavior.

## Feature Engineering

New features were created to better capture financial behavior and credit risk patterns.

```python
# Feature Engineering

# Debt to income ratio (financial pressure indicator)
df['debt_to_income'] = df['DebtRatio']

# Income per dependent (household burden)
df['income_per_person'] = df['MonthlyIncome'] / (df['NumberOfDependents'] + 1)

# Total delinquency count
df['total_delinquency'] = (
    df['NumberOfTime30-59DaysPastDueNotWorse'] +
    df['NumberOfTimes90DaysLate'] +
    df['NumberOfTime60-89DaysPastDueNotWorse']
)

# Credit lines per person
df['credit_lines_per_person'] = df['NumberOfOpenCreditLinesAndLoans'] / (df['NumberOfDependents'] + 1)
```

These features help capture:

- Financial stress (debt vs income)
- Household burden
- Payment behavior
- Credit exposure

These are key drivers in real-world credit risk models.

## Feature Insights

The engineered features reveal important patterns:

- Most customers show low delinquency levels, but a small group presents extremely high values, indicating high-risk behavior.
- Debt-to-income ratio varies significantly, highlighting differences in financial stress across customers.
- Income distribution shows potential anomalies (e.g., zero income), which may require further validation.

These features provide strong signals for credit risk segmentation.

## Risk Scoring Model

A simple rule-based scoring system was developed to classify customers based on their credit risk.

```python
# Initialize score
df['risk_score'] = 0

# High debt burden
df.loc[df['debt_to_income'] > 0.4, 'risk_score'] += 2

# Low income per person
df.loc[df['income_per_person'] < 2000, 'risk_score'] += 2

# Delinquency behavior
df.loc[df['total_delinquency'] > 0, 'risk_score'] += 3

# High credit utilization
df.loc[df['RevolvingUtilizationOfUnsecuredLines'] > 0.5, 'risk_score'] += 2
```

```python
# Risk categories
def classify_risk(score):
    if score <= 2:
        return "Low Risk"
    elif score <= 5:
        return "Medium Risk"
    else:
        return "High Risk"

df['risk_category'] = df['risk_score'].apply(classify_risk)
```

```python
df['risk_category'].value_counts()
```

## Risk Segmentation

The scoring model classified customers into three risk categories:

- 64% Low Risk
- 26% Medium Risk
- 9% High Risk

Most customers fall into the low-risk category, indicating a generally healthy portfolio. However, a smaller segment represents high-risk clients with significantly higher probability of default.

## Risk Profile Analysis

Clear differences emerge between risk segments:

- High Risk customers show lower income, higher debt levels, and significantly higher delinquency.
- Low Risk customers demonstrate strong financial stability, with higher income and no delinquency.

These patterns confirm that financial behavior and repayment history are key drivers of credit risk.

## Business Recommendations

Based on the analysis, the following strategies are recommended:

- Limit loan approvals for high-risk customers or apply stricter conditions (e.g., lower loan amounts, higher interest rates)
- Prioritize low-risk customers for loan expansion strategies
- Develop targeted policies for medium-risk customers to maximize growth while managing risk
- Implement early warning systems based on delinquency behavior

These actions can help reduce default rates while maintaining portfolio growth.
